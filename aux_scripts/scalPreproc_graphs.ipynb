# Cordis Figures

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
num_workers = [1, 2, 4, 8, 16, 32]
cleaning_time = [187.75, 98.93, 56.15, 31.38, 22.61, 19.24]
export_time = np.array([198.77, 101.36, 57.77, 34.44, 25.56, 23.95])
```

```python
plt.rcParams["figure.figsize"] = [8, 6]
plt.rcParams["figure.autolayout"] = True
fig, ax = plt.subplots()

plt.loglog(num_workers, cleaning_time, color='red', marker='d',
     linewidth=2, markersize=12, fillstyle='full', linestyle='dashed')
plt.plot([1,32], [19.55, 19.55], linewidth=2)
plt.xlabel("Number of Workers", fontsize=18)
plt.ylabel("Cleaning Time (seconds)", fontsize=18)
plt.legend(["User selected num_workers", "Default num_workers"], fontsize=16)
ax.set_xticks(num_workers)
ax.set_xticklabels([str(el) for el in num_workers])
ax.set_yticks(cleaning_time)
ax.set_yticklabels([str(el) for el in cleaning_time])
ax.tick_params(axis='both', which='major', labelsize=16)
plt.savefig('cordis_cleaning.png')
```

```python
plt.rcParams["figure.figsize"] = [8, 6]
plt.rcParams["figure.autolayout"] = True
fig, ax = plt.subplots()

plt.loglog(num_workers, export_time, color='green', marker='o', linestyle='dashed',
     linewidth=2, markersize=12, fillstyle='full')
plt.plot([1,32], [27.51, 27.51], linewidth=2)
plt.xlabel("Number of Workers", fontsize=18)
plt.ylabel("Export Time (seconds)", fontsize=18)
plt.legend(["User selected num_workers", "Default num_workers"], fontsize=16)
ax.set_xticks(num_workers)
ax.set_xticklabels([str(el) for el in num_workers])
ax.set_yticks(cleaning_time)
ax.set_yticklabels([str(el) for el in export_time])
ax.tick_params(axis='both', which='major', labelsize=16)
plt.savefig("cordis_export.png") 
```

## CPU and Memory usage evolution

```python
import pandas as pd
mem_use = pd.read_csv("../scalPreproc_nw4/S2CS_10/stats/mem_use.txt")
```

```python
seconds = np.arange(len(mem_use))
mem_use = mem_use[::2]
seconds = seconds[::2]
```

```python
plt.rcParams["figure.figsize"] = [8, 8]
plt.rcParams["figure.autolayout"] = True
fig, (ax1, ax2) = plt.subplots(2, 1, gridspec_kw={'height_ratios': [3, 1]})

ax1.plot(seconds, [el/1024 for el in mem_use["rss"].values.tolist()], "r")
ax1.plot(seconds, [el/1024 for el in mem_use["vms"].values.tolist()])
ax2.plot(seconds, [el/100 for el in mem_use["cpu"].values.tolist()], "g")

ax1.set_xticks([])
ax2.set_xlabel("Seconds", fontsize=18)
ax1.set_ylabel("Memory Used (GB)", fontsize=18)
ax2.set_ylabel("Number of CPUs", fontsize=18)
ax1.tick_params(axis='both', which='major', labelsize=16)
ax2.tick_params(axis='both', which='major', labelsize=16)
ax2.set_ylim([0,8])
ax1.legend(["RAM memory", "Virtual memory", "CPUs"], fontsize=16)
ax1.set_title("S2CS (10%) - num_workers = 4", fontsize=20)
fig.savefig("./scal_graphs/S2CS10_nw4.png")
```

# Overall preprocessing time

```python
total_time = []
n_workers = [4, 8, 16, 32, 0]

for k in [4, 8, 16, 32, 0]:
    k_times = []
    for sz in dts_size:
        mem_use = pd.read_csv("../scalPreproc_nw" + str(k) + "/S2CS_" + str(sz) + "/stats/mem_use.txt")
        k_times.append(len(mem_use))
    total_time.append(k_times)
    
total_time = np.array(total_time)
```

```python
plt.rcParams["figure.figsize"] = [8, 6]
plt.rcParams["figure.autolayout"] = True
fig, ax = plt.subplots()

plt.semilogx(dts_size, total_time.T, marker='.',
     linewidth=2, markersize=16, fillstyle='full', linestyle='dashed')

plt.xlabel("Sampling rate (%)", fontsize=18)
plt.ylabel("Total preprocessing time (s)", fontsize=18)
plt.legend(["n workers = " + str(k) for k in [4, 8, 16, 32, 0]], fontsize=16)
ax.set_xticks(dts_size)
ax.set_xticklabels([str(el) for el in dts_size])
#ax.set_yticks(cleaning_time)
#ax.set_yticklabels([str(el) for el in cleaning_time])
ax.tick_params(axis='both', which='major', labelsize=16)

plt.savefig('scal_graphs/total_preproc_time.png')
```

```python
plt.rcParams["figure.figsize"] = [8, 6]
plt.rcParams["figure.autolayout"] = True
fig, ax = plt.subplots()

plt.semilogx(dts_size, total_time.T-dictio_times.T, marker='.',
     linewidth=2, markersize=16, fillstyle='full', linestyle='dashed')

plt.xlabel("Sampling rate (%)", fontsize=18)
plt.ylabel("Cleaning + Export Times (s)", fontsize=18)
plt.legend(["n workers = " + str(k) for k in [4, 8, 16, 32, 0]], fontsize=16)
ax.set_xticks(dts_size)
ax.set_xticklabels([str(el) for el in dts_size])
#ax.set_yticks(cleaning_time)
#ax.set_yticklabels([str(el) for el in cleaning_time])
ax.tick_params(axis='both', which='major', labelsize=16)

plt.savefig('scal_graphs/total_dask_time.png')
```

# Spark

```python
spark_times = np.array([[1576, 2092, 3399], [284, 818, 2115], [234, 619, 1525], [210, 528, 1692]])
dts_size = [10, 30, 100]
cores_node = [2, 4, 8, 16]
```

```python
plt.rcParams["figure.figsize"] = [8, 6]
plt.rcParams["figure.autolayout"] = True
fig, ax = plt.subplots()

plt.semilogx(dts_size, spark_times.T, marker='.',
     linewidth=2, markersize=16, fillstyle='full', linestyle='dashed')

plt.xlabel("Sampling rate (%)", fontsize=18)
plt.ylabel("Total preprocessing time (s)", fontsize=18)
plt.legend([str(k) + " cores per node" for k in cores_node], fontsize=16)
ax.set_xticks(dts_size)
ax.set_xticklabels([str(el) for el in dts_size])
#ax.set_yticks(cleaning_time)
#ax.set_yticklabels([str(el) for el in cleaning_time])
ax.tick_params(axis='both', which='major', labelsize=16)
ax.set_title('Preprocessing with a Spark cluster with 10 nodes', fontsize=20)

plt.savefig('scal_graphs/spark_time.png')
```
